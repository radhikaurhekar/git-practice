```python
inventory = {
    "t-shirt": 5,
    "mug": 4,
    "hat": 6,
    "book": 3,
    "keychain": 7
}

customer_orders = {"t-shirt", "mug", "hat"}
```

```python
inventory[list(customer_orders)[0]] -= 1
inventory[list(customer_orders)[1]] -= 1
inventory[list(customer_orders)[2]] -= 1
```

```python
print("t-shirt:", inventory["t-shirt"])
print("mug:", inventory["mug"])
print("hat:", inventory["hat"])
print("book:", inventory["book"])
print("keychain:", inventory["keychain"])
```
